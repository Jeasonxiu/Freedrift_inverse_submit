```python
# constants
rhoa = 1.25 
rhoi = 917.0 
rhoo = 1026.0 
fc   = 1.46e-4 
hice_1 = 1.0e-0
hifc_1 = hice_1*fc
hice_2 = 1.1e-0
hifc_2 = hice_2*fc
hice_3 = 0.9e-0
hifc_3 = hice_3*fc
hice_4 = 0.9-0
hifc_4 = hice_4*fc
hice_5 = 1.0e-0
hifc_5 = hice_5*fc
```

```python
# set up wind and ocean
wind_mag_1 = 15.0
wind_ang_1 = np.deg2rad(0)
uwind_1    = wind_mag_1*np.cos(wind_ang_1)
vwind_1    = wind_mag_1*np.sin(wind_ang_1)
ocn_mag_1  = 0.05
ocn_ang_1  = np.deg2rad(10)
uocn_1     = ocn_mag_1*np.cos(ocn_ang_1)
vocn_1     = ocn_mag_1*np.sin(ocn_ang_1)

wind_mag_2 = 12.0
wind_ang_2 = np.deg2rad(5)
uwind_2    = wind_mag_2*np.cos(wind_ang_2)
vwind_2    = wind_mag_2*np.sin(wind_ang_2)
ocn_mag_2  = 0.04
ocn_ang_2  = np.deg2rad(10)
uocn_2     = ocn_mag_2*np.cos(ocn_ang_2)
vocn_2     = ocn_mag_2*np.sin(ocn_ang_2)

wind_mag_3 = 16.0
wind_ang_3 = np.deg2rad(10)
uwind_3    = wind_mag_3*np.cos(wind_ang_3)
vwind_3    = wind_mag_3*np.sin(wind_ang_3)
ocn_mag_3  = 0.03
ocn_ang_3  = np.deg2rad(13)
uocn_3     = ocn_mag_3*np.cos(ocn_ang_3)
vocn_3     = ocn_mag_3*np.sin(ocn_ang_3)

wind_mag_4 = 12.0
wind_ang_4 = np.deg2rad(12)
uwind_4    = wind_mag_4*np.cos(wind_ang_4)
vwind_4    = wind_mag_4*np.sin(wind_ang_4)
ocn_mag_4  = 0.003
ocn_ang_4  = np.deg2rad(15)
uocn_4     = ocn_mag_4*np.cos(ocn_ang_4)
vocn_4     = ocn_mag_4*np.sin(ocn_ang_4)

wind_mag_5 = 17.0
wind_ang_5 = np.deg2rad(1)
uwind_5    = wind_mag_5*np.cos(wind_ang_5)
vwind_5    = wind_mag_5*np.sin(wind_ang_5)
ocn_mag_5  = 0.0003
ocn_ang_5  = np.deg2rad(15)
uocn_5     = ocn_mag_5*np.cos(ocn_ang_5)
vocn_5     = ocn_mag_5*np.sin(ocn_ang_5)

Ca_min = 5.0e-4
Ca_max = 9.0e-3
Co_min = 1.0e-3
Co_max = 3.0e-2

# invert for thO and thA-thO
# these replace for thO
thA_min = np.deg2rad(-10)
thA_max = np.deg2rad(45)
# these replace for thA-thO
thO_min = np.deg2rad(-30)
thO_max = np.deg2rad(30)

Ca     = 4e-3
ThetaA = np.deg2rad(0)
Co     = 5e-3
ThetaO = np.deg2rad(20) # original
```

```python
# solution constants etc
def nansen(Ca,Co):
    return np.sqrt((rhoa*Ca)/(rhoo*Co))

def nansen2_Ua(Ca,Co,Ua):
    return ((rhoa*Ca)/(rhoo*Co))*Ua

def rossby_p(Co,hifc):
    return rhoi*hifc/(rhoo*Co)
# print(rossby_p(Co)/Na/wind_mag)
nansen_use = nansen(Ca,Co)**2
rossby_use = rossby_p(Co,hifc_1)/hifc_1
print(nansen(Ca_min,Co_max),nansen(Ca,Co),nansen(Ca_max,Co_min))
print(nansen2_Ua(Ca_min,Co_max,wind_mag_1),nansen2_Ua(Ca,Co,wind_mag_1),
      nansen2_Ua(Ca_max,Co_min,wind_mag_1))
print(rossby_p(Co_max,hifc_1),rossby_p(Co,hifc_1),rossby_p(Co_min,hifc_1))
```

```python
# solve ice_mag 1
uwind    = uwind_1
vwind    = vwind_1
wind_mag = wind_mag_1
uocn     = uocn_1
vocn     = vocn_1
ocn_mag  = ocn_mag_1
hifc     = hifc_1
Eq11 = np.tan(ThetaO)
Eq12 = (rossby_p(Co,hifc)/wind_mag)
Eq13 = np.cos(ThetaO)

Eq21 = 2*np.sin(ThetaO)*rossby_p(Co,hifc)/wind_mag
Eq22 = (rossby_p(Co,hifc)**2)/(wind_mag**2)
Eq23 = nansen(Ca,Co)**4

y1 = nansen(Ca,Co)
yi = y1
eps = 1.0e-4
conv = 0
while conv == 0:
    fi =   yi**4 +   Eq21*(yi**3) +   Eq22*(yi**2) - Eq23
    dfi= 4*yi**3 + 3*Eq21*(yi**2) + 2*Eq22*yi
    yi = yi - fi/dfi
    if (np.abs(fi/dfi) < eps):
        conv = 1
# x = ThetaA + np.arctan(Eq11 + Eq12/(yi*Eq13))
x = - ThetaA + np.arctan(Eq11 + Eq12/(yi*Eq13))
alpha = yi
ThetaI = x

uice = alpha*(np.cos(-ThetaI)*uwind - np.sin(-ThetaI)*vwind) + uocn
vice = alpha*(np.cos(-ThetaI)*vwind + np.sin(-ThetaI)*uwind) + vocn
ice_mag = np.hypot(uice,vice)

uice_1    = uice
vice_1    = vice
ice_mag_1 = ice_mag 
print(ice_mag)
```

```python
# solve ice_mag 2
uwind    = uwind_2
vwind    = vwind_2
wind_mag = wind_mag_2
uocn     = uocn_2
vocn     = vocn_2
ocn_mag  = ocn_mag_2
hifc     = hifc_2
Eq11 = np.tan(ThetaO)
Eq12 = (rossby_p(Co,hifc)/wind_mag)
Eq13 = np.cos(ThetaO)

Eq21 = 2*np.sin(ThetaO)*rossby_p(Co,hifc)/wind_mag
Eq22 = (rossby_p(Co,hifc)**2)/(wind_mag**2)
Eq23 = nansen(Ca,Co)**4

y1 = nansen(Ca,Co)
yi = y1
eps = 1.0e-4
conv = 0
while conv == 0:
    fi =   yi**4 +   Eq21*(yi**3) +   Eq22*(yi**2) - Eq23
    dfi= 4*yi**3 + 3*Eq21*(yi**2) + 2*Eq22*yi
    yi = yi - fi/dfi
    if (np.abs(fi/dfi) < eps):
        conv = 1
# x = ThetaA + np.arctan(Eq11 + Eq12/(yi*Eq13))
x = - ThetaA + np.arctan(Eq11 + Eq12/(yi*Eq13))
alpha = yi
ThetaI = x

uice = alpha*(np.cos(-ThetaI)*uwind - np.sin(-ThetaI)*vwind) + uocn
vice = alpha*(np.cos(-ThetaI)*vwind + np.sin(-ThetaI)*uwind) + vocn
ice_mag = np.hypot(uice,vice)

uice_2    = uice
vice_2    = vice
ice_mag_2 = ice_mag 
print(ice_mag)
```

```python
# solve ice_mag 4
uwind    = uwind_4
vwind    = vwind_4
wind_mag = wind_mag_4
uocn     = uocn_4
vocn     = vocn_4
ocn_mag  = ocn_mag_4
hifc     = hifc_4
Eq11 = np.tan(ThetaO)
Eq12 = (rossby_p(Co,hifc)/wind_mag)
Eq13 = np.cos(ThetaO)

Eq21 = 2*np.sin(ThetaO)*rossby_p(Co,hifc)/wind_mag
Eq22 = (rossby_p(Co,hifc)**2)/(wind_mag**2)
Eq23 = nansen(Ca,Co)**4

y1 = nansen(Ca,Co)
yi = y1
eps = 1.0e-4
conv = 0
while conv == 0:
    fi =   yi**4 +   Eq21*(yi**3) +   Eq22*(yi**2) - Eq23
    dfi= 4*yi**3 + 3*Eq21*(yi**2) + 2*Eq22*yi
    yi = yi - fi/dfi
    if (np.abs(fi/dfi) < eps):
        conv = 1
# x = ThetaA + np.arctan(Eq11 + Eq12/(yi*Eq13))
x = - ThetaA + np.arctan(Eq11 + Eq12/(yi*Eq13))
alpha = yi
ThetaI = x

uice = alpha*(np.cos(-ThetaI)*uwind - np.sin(-ThetaI)*vwind) + uocn
vice = alpha*(np.cos(-ThetaI)*vwind + np.sin(-ThetaI)*uwind) + vocn
ice_mag = np.hypot(uice,vice)

uice_4    = uice
vice_4    = vice
ice_mag_4 = ice_mag 
print(ice_mag)
```

```python
# solve ice_mag 5
uwind    = uwind_5
vwind    = vwind_5
wind_mag = wind_mag_5
uocn     = uocn_5
vocn     = vocn_5
ocn_mag  = ocn_mag_5
hifc     = hifc_5
Eq11 = np.tan(ThetaO)
Eq12 = (rossby_p(Co,hifc)/wind_mag)
Eq13 = np.cos(ThetaO)

Eq21 = 2*np.sin(ThetaO)*rossby_p(Co,hifc)/wind_mag
Eq22 = (rossby_p(Co,hifc)**2)/(wind_mag**2)
Eq23 = nansen(Ca,Co)**4

y1 = nansen(Ca,Co)
yi = y1
eps = 1.0e-4
conv = 0
while conv == 0:
    fi =   yi**4 +   Eq21*(yi**3) +   Eq22*(yi**2) - Eq23
    dfi= 4*yi**3 + 3*Eq21*(yi**2) + 2*Eq22*yi
    yi = yi - fi/dfi
    if (np.abs(fi/dfi) < eps):
        conv = 1
# x = ThetaA + np.arctan(Eq11 + Eq12/(yi*Eq13))
x = - ThetaA + np.arctan(Eq11 + Eq12/(yi*Eq13))
alpha = yi
ThetaI = x

uice = alpha*(np.cos(-ThetaI)*uwind - np.sin(-ThetaI)*vwind) + uocn
vice = alpha*(np.cos(-ThetaI)*vwind + np.sin(-ThetaI)*uwind) + vocn
ice_mag = np.hypot(uice,vice)

uice_5    = uice
vice_5    = vice
ice_mag_5 = ice_mag 
print(ice_mag)
```

```python
# now try back in the mtm balance
u_ice_ocn = uocn - uice
v_ice_ocn = vocn - vice
ice_ocn_mag = np.hypot(u_ice_ocn,v_ice_ocn)

atmo_stressx = nansen2_Ua(Ca,Co,wind_mag)*( uwind*np.cos(ThetaA) - vwind*np.sin(ThetaA) )
atmo_stressy = nansen2_Ua(Ca,Co,wind_mag)*( vwind*np.cos(ThetaA) + uwind*np.sin(ThetaA) )

ocn_stressx  = ice_ocn_mag*( u_ice_ocn*np.cos(ThetaO) - v_ice_ocn*np.sin(ThetaO) )
ocn_stressy  = ice_ocn_mag*( v_ice_ocn*np.cos(ThetaO) + u_ice_ocn*np.sin(ThetaO) )

cor_tiltx = -v_ice_ocn*rossby_p(Co,hifc)
cor_tilty =  u_ice_ocn*rossby_p(Co,hifc)

tot_mtm_x = atmo_stressx + ocn_stressx + cor_tiltx
tot_mtm_y = atmo_stressy + ocn_stressy + cor_tilty

print(np.hypot(tot_mtm_x,tot_mtm_y))
print(tot_mtm_x)
print(tot_mtm_y)
```
