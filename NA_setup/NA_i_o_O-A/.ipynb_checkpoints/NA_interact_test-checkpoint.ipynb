```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
# from scipy.io import FortranFile
# import struct
```

```python
# constants
rhoa = 1.25 
rhoi = 917.0 
rhoo = 1026.0 
fc   = 1.46e-4 
hice = 1.0e-0
hifc = hice*fc
```

```python
# set up wind and ocean
wind_mag = 9.0
wind_ang = np.deg2rad(0)
uwind    = wind_mag*np.cos(wind_ang)
vwind    = wind_mag*np.sin(wind_ang)
ocn_mag  = 0.05
ocn_ang  = np.deg2rad(80)
uocn     = ocn_mag*np.cos(ocn_ang)
vocn     = ocn_mag*np.sin(ocn_ang)

Ca_min = 0.0e-3
Ca_max = 3.0e-3
Co_min = 1.0e-3
Co_max = 3.0e-2

Ca     = 1e-3
ThetaA = np.deg2rad(15)
Co     = 5e-3
ThetaO = np.deg2rad(25)
```

```python
# solution constants etc
def nansen(Ca,Co):
    return np.sqrt((rhoa*Ca)/(rhoo*Co))

def rossby_p(Co):
    return rhoi*hifc/(rhoo*Co)
# print(rossby_p(Co)/Na/wind_mag)
nansen_use = nansen(Ca,Co)
rossby_use = rossby_p(Co)
print(nansen(Ca_min,Co_max),nansen(Ca,Co),nansen(Ca_max,Co_min))
print(rossby_p(Co_max),rossby_p(Co),rossby_p(Co_min))
```

```python
# solve ice_mag
Eq11 = np.tan(ThetaO)
Eq12 = (rossby_p(Co)/wind_mag)
Eq13 = np.cos(ThetaO)

Eq21 = 2*np.sin(ThetaO)*rossby_p(Co)/wind_mag
Eq22 = (rossby_p(Co)**2)/(wind_mag**2)
Eq23 = nansen(Ca,Co)**4

y1 = nansen(Ca,Co)
yi = y1
eps = 1.0e-4
conv = 0
while conv == 0:
    fi =   yi**4 +   Eq21*(yi**3) +   Eq22*(yi**2) - Eq23
    dfi= 4*yi**3 + 3*Eq21*(yi**2) + 2*Eq22*yi
    yi = yi - fi/dfi
    if (np.abs(fi/dfi) < eps):
        conv = 1
# x = ThetaA + np.arctan(Eq11 + Eq12/(yi*Eq13))
x = - ThetaA + np.arctan(Eq11 + Eq12/(yi*Eq13))
alpha = yi
ThetaI = x

uice = alpha*(np.cos(-ThetaI)*uwind - np.sin(-ThetaI)*vwind) + uocn
vice = alpha*(np.cos(-ThetaI)*vwind + np.sin(-ThetaI)*uwind) + vocn
ice_mag = np.hypot(uice,vice)

u_ice_ocn = uocn - uice
v_ice_ocn = vocn - vice 
ice_ocn_mag = np.hypot(u_ice_ocn,v_ice_ocn)
print(ice_mag,ice_ocn_mag)
```

```python
# now try back in the mtm balance
atmo_stressx = nansen(Ca,Co)**2*wind_mag*( uwind*np.cos(ThetaA) - vwind*np.sin(ThetaA) )
atmo_stressy = nansen(Ca,Co)**2*wind_mag*( vwind*np.cos(ThetaA) + uwind*np.sin(ThetaA) )

ocn_stressx  = ice_ocn_mag*( u_ice_ocn*np.cos(ThetaO) - v_ice_ocn*np.sin(ThetaO) )
ocn_stressy  = ice_ocn_mag*( v_ice_ocn*np.cos(ThetaO) + u_ice_ocn*np.sin(ThetaO) )

cor_tiltx = -v_ice_ocn*rossby_p(Co)
cor_tilty =  u_ice_ocn*rossby_p(Co)

tot_mtm_x = atmo_stressx + ocn_stressx + cor_tiltx
tot_mtm_y = atmo_stressy + ocn_stressy + cor_tilty

print(np.hypot(tot_mtm_x,tot_mtm_y))
print(tot_mtm_x)
print(tot_mtm_y)
```
