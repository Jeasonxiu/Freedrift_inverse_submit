```python
# constants
rhoa = 1.25 
rhoi = 917.0 
rhoo = 1026.0 
fc   = 1.46e-4 
hice_1 = 1.0e-0
hifc_1 = hice_1*fc
hice_2 = 1.1e-0
hifc_2 = hice_2*fc
hice_3 = 0.9e-0
hifc_3 = hice_3*fc
hice_4 = 0.9-0
hifc_4 = hice_4*fc
hice_5 = 1.0e-0
hifc_5 = hice_5*fc
```

```python
# set up wind and ocean
wind_mag_1 = 5.0
wind_ang_1 = np.deg2rad(0)
uwind_1    = wind_mag_1*np.cos(wind_ang_1)
vwind_1    = wind_mag_1*np.sin(wind_ang_1)
ocn_mag_1  = 0.1
ocn_ang_1  = np.deg2rad(10)
uocn_1     = ocn_mag_1*np.cos(ocn_ang_1)
vocn_1     = ocn_mag_1*np.sin(ocn_ang_1)

wind_mag_2 = 2.0
wind_ang_2 = np.deg2rad(5)
uwind_2    = wind_mag_2*np.cos(wind_ang_2)
vwind_2    = wind_mag_2*np.sin(wind_ang_2)
ocn_mag_2  = 0.2
ocn_ang_2  = np.deg2rad(-40)
uocn_2     = ocn_mag_2*np.cos(ocn_ang_2)
vocn_2     = ocn_mag_2*np.sin(ocn_ang_2)

wind_mag_3 = 6.0
wind_ang_3 = np.deg2rad(10)
uwind_3    = wind_mag_3*np.cos(wind_ang_3)
vwind_3    = wind_mag_3*np.sin(wind_ang_3)
ocn_mag_3  = 0.06
ocn_ang_3  = np.deg2rad(63)
uocn_3     = ocn_mag_3*np.cos(ocn_ang_3)
vocn_3     = ocn_mag_3*np.sin(ocn_ang_3)

wind_mag_4 = 2.0
wind_ang_4 = np.deg2rad(12)
uwind_4    = wind_mag_4*np.cos(wind_ang_4)
vwind_4    = wind_mag_4*np.sin(wind_ang_4)
ocn_mag_4  = 0.03
ocn_ang_4  = np.deg2rad(15)
uocn_4     = ocn_mag_4*np.cos(ocn_ang_4)
vocn_4     = ocn_mag_4*np.sin(ocn_ang_4)

wind_mag_5 = 7.0
wind_ang_5 = np.deg2rad(1)
uwind_5    = wind_mag_5*np.cos(wind_ang_5)
vwind_5    = wind_mag_5*np.sin(wind_ang_5)
ocn_mag_5  = 0.1
ocn_ang_5  = np.deg2rad(15)
uocn_5     = ocn_mag_5*np.cos(ocn_ang_5)
vocn_5     = ocn_mag_5*np.sin(ocn_ang_5)

Ca_min = 1.0e-4
Ca_max = 5.0e-3
Co_min = 2.0e-3
Co_max = 6.0e-2

# invert for thO and thA-thO
# these replace for thO
thA_min = np.deg2rad(-10)
thA_max = np.deg2rad(45)
# these replace for thA-thO
thO_min = np.deg2rad(-30)
thO_max = np.deg2rad(30)

Ca     = 1e-3
ThetaA = np.deg2rad(5)
Co     = 4e-3
ThetaO = np.deg2rad(20) # original
```

```python
nansen_use = freedrift.nansen(Ca,Co)**2
rossby_use = freedrift.rossby_pp(Co)
print(freedrift.nansen(Ca_min,Co_max),freedrift.nansen(Ca,Co),freedrift.nansen(Ca_max,Co_min))
print(freedrift.nansen2_Ua(Ca_min,Co_max,wind_mag_1),freedrift.nansen2_Ua(Ca,Co,wind_mag_1),
      freedrift.nansen2_Ua(Ca_max,Co_min,wind_mag_1))
print(freedrift.rossby_p(Co_max,hifc_1),freedrift.rossby_p(Co,hifc_1),freedrift.rossby_p(Co_min,hifc_1))
print(freedrift.rossby_pp(Co_max),freedrift.rossby_pp(Co),freedrift.rossby_pp(Co_min))
```

```python
# put noise on all the parameters before solving
sig_c = 0.2 # plus/minus lever for random error
# problem 1
na2_1 = freedrift.nansen(Ca*np.random.normal(loc=1,scale=sig_c),Co*np.random.normal(loc=1,scale=sig_c))**2
rop_1 = freedrift.rossby_pp(Co*np.random.normal(loc=1,scale=sig_c))
thA_1 = ThetaA*np.random.normal(loc=1,scale=sig_c)
thO_1 = ThetaO*np.random.normal(loc=1,scale=sig_c)
uice_1,vice_1 = freedrift.ice_mag(uwind_1,vwind_1,uocn_1,vocn_1,
                 na2_1,rop_1*hifc_1,thA_1,thO_1)
ice_mag_1     = np.hypot(uice_1,vice_1)
ice_ocn_mag_1 = np.hypot(uocn_1 - uice_1,vocn_1 - vice_1)
# problem 2
na2_2 = freedrift.nansen(Ca*np.random.normal(loc=1,scale=sig_c),Co*np.random.normal(loc=1,scale=sig_c))**2
rop_2 = freedrift.rossby_pp(Co*np.random.normal(loc=1,scale=sig_c))
thA_2 = ThetaA*np.random.normal(loc=1,scale=sig_c)
thO_2 = ThetaO*np.random.normal(loc=1,scale=sig_c)
uice_2,vice_2 = freedrift.ice_mag(uwind_2,vwind_2,uocn_2,vocn_2,
                 na2_2,rop_2*hifc_2,thA_2,thO_2)
ice_mag_2     = np.hypot(uice_2,vice_2)
ice_ocn_mag_2 = np.hypot(uocn_2 - uice_2,vocn_2 - vice_2)
# problem 3
na2_3 = freedrift.nansen(Ca*np.random.normal(loc=1,scale=sig_c),Co*np.random.normal(loc=1,scale=sig_c))**2
rop_3 = freedrift.rossby_pp(Co*np.random.normal(loc=1,scale=sig_c))
thA_3 = ThetaA*np.random.normal(loc=1,scale=sig_c)
thO_3 = ThetaO*np.random.normal(loc=1,scale=sig_c)
uice_3,vice_3 = freedrift.ice_mag(uwind_3,vwind_3,uocn_3,vocn_3,
                 na2_3,rop_3*hifc_3,thA_3,thO_3)
ice_mag_3     = np.hypot(uice_3,vice_3)
ice_ocn_mag_3 = np.hypot(uocn_3 - uice_3,vocn_3 - vice_3)
# problem 4
na2_4 = freedrift.nansen(Ca*np.random.normal(loc=1,scale=sig_c),Co*np.random.normal(loc=1,scale=sig_c))**2
rop_4 = freedrift.rossby_pp(Co*np.random.normal(loc=1,scale=sig_c))
thA_4 = ThetaA*np.random.normal(loc=1,scale=sig_c)
thO_4 = ThetaO*np.random.normal(loc=1,scale=sig_c)
uice_4,vice_4 = freedrift.ice_mag(uwind_4,vwind_4,uocn_4,vocn_4,
                 na2_4,rop_4*hifc_4,thA_4,thO_4)
ice_mag_4     = np.hypot(uice_4,vice_4)
ice_ocn_mag_4 = np.hypot(uocn_4 - uice_4,vocn_4 - vice_4)
# problem 5
na2_5 = freedrift.nansen(Ca*np.random.normal(loc=1,scale=sig_c),Co*np.random.normal(loc=1,scale=sig_c))**2
rop_5 = freedrift.rossby_pp(Co*np.random.normal(loc=1,scale=sig_c))
thA_5 = ThetaA*np.random.normal(loc=1,scale=sig_c)
thO_5 = ThetaO*np.random.normal(loc=1,scale=sig_c)
uice_5,vice_5 = freedrift.ice_mag(uwind_5,vwind_5,uocn_5,vocn_5,
                 na2_5,rop_5*hifc_5,thA_5,thO_5)
ice_mag_5     = np.hypot(uice_5,vice_5)
ice_ocn_mag_5 = np.hypot(uocn_5 - uice_5,vocn_5 - vice_5)
```
